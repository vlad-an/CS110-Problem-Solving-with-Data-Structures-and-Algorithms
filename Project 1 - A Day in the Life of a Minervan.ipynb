## ***Q2 Preparing your algorithmic strategy***
**A**

A priority queue is a well-suited data structure to prioritize tasks because it allows for efficient insertion and retrieval of elements based on their priority. Each element of a priority queue is associated with a priority value, and the queue is designed to always return the element with the highest priority first. With a priority queue, we can easily store any information associated with each instance of a task, which allows us easily access that information about a task at any position in a queue. In addition, we can manipulate the queue/heap by adding tasks, deleting tasks, changing priority values, and so on without reconstructing the whole heap, as its properties and methods can ensure that the task with the highest priority is always at the top of the queue. 

This makes this data structure ideal for situations where tasks have different priorities and need to be processed in a specific order. For example, in a scheduling system, tasks with a higher priority (e.g., urgent tasks) need to be executed before tasks with a lower priority (e.g., non-urgent tasks).

Sorting a list of tasks can be an option, but it can be **inefficient** if the list is large and frequently changing. Sorting a list requires O(n log n) / O(n^2) time complexity depending on the type of sorting algorithm used, while a priority queue can perform insertion and retrieval operations in O(log n) time complexity. This means that the algorithm's running time grows in proportion to the logarithm of the input size n. Moreover, sorting a list requires **re-sorting** every time a new task is added (or other manipulation performed), which is not optimal for tasks with dynamic priorities. By the contrast, using priority queues, we can insert/add element one by one by moving them to their correct positions. We don't need to re-sort every time.

**Priority Queue Time Complexity Analysis**

We can use heap, a specific type of priority queue, as example. 

- Insertion:
To insert an element into a priority queue, we add it to the bottom of the binary heap and then perform a heapify operation. This operation moves the element up the tree until it is in its correct position relative to its parent nodes. The time complexity of insertion in a binary heap-based priority queue is **O(log n)**, where n is the number of elements in the priority queue.

- Deletion:
To remove the element with the highest priority from a priority queue, we remove the root node of the binary heap, which contains the highest-priority element. We then replace the root node with the last element in the heap and perform a heapify operation to maintain the heap property. The time complexity of deletion in a binary heap-based priority queue is also **O(log n)**.

- Sorting:
To sort a collection of elements using a priority queue, we insert each element into the priority queue in **O(log n) time**, resulting in a binary heap-based priority queue with all elements sorted by priority. We can then remove each element from the priority queue in O(log n) time, resulting in a sorted list of elements. The time complexity of sorting using a priority queue is **O(n log n)**, where n is the number of elements in the collection.

To accommodate the fixed-time and flexible-time tasks, we would need to create two separate priority queues. First, we would store the tasks with flexible times where their order would be determined by their **priority**: highest priority tasks are at the top of the queue. In the second queue, we order the fixed-time tasks based on the **execution time**: earlier tasks come first in the queue. The scheduler would work by taking the highest priority task from the flexible-time tasks queue. If the timing or duration of this task does not interfere with the earliest fixed-time task, we execute it. Otherwise, we do not perform a task until the time for the earliest fixed-time task comes, at which point we execute this fixed-time task. After that, we repeat the procedure until all tasks are complete. In other words, we are trying to make sure none of the flexible-time tasks interfere with the timing of the fixed-time tasks.

**Detailed Explanation**

If we are explaining the algorithm to an older audience (e.g., a peer) but has not taken CS110 before:<br>

To start with, we need to define what a task is. In our case, it is an object that has the following characteristics: id number, description, duration (in minutes), starting time (in minutes, if the task must happen at fixed-time), and a list of other tasks that the task depends on (meaning must be executed after the tasks it depends on are executed). 

Then, we need a way to store these tasks in proper order. We will build 2 queues for this purpose where queue one would store the flexible-time tasks and order them based on their priority values (highest priority - at the front of the queue to be executed earlier), and queue two would store fixed-time tasks and order them based on their scheduled starting time (earlier time - at the front of the queue to be executed earlier).

Lastly, we need a procedure to decide which task to execute at a given moment, execute it, and move on to the next task until all tasks are executed. We would start by assessing the highest priority task from the flexible-time tasks queue and the earliest task from the fixed-time tasks queue. If the starting time and the starting time plus the duration of the highest priority flexible-time task do not interfere with the starting time of the earliest fixed-time task, then we execute the highest priority flexible-time task. We keep taking and executing the highest priority flexible-time tasks as long as they start and finish before the start of the fixed-time task. When we end a certain flexible-time task but can not start a new flexible-time task due to fixed-time task interference, we simply wait until the fixed-time task needs to be executed and execute it then. This way, all fixed-time tasks will be executed at scheduled times, and flexible-time tasks will be scheduled at times that do not interfere with fixed-time tasks in any way and will be executed based on their priority.

The complete mathematical process of deciding and calculating the **priority value** of flexible tasks is detailed in the next section, C.

**C**

For our task scheduler, we define the priority of a task as a function of its duration and the number of dependencies. Intuitively, the longer the duration of a task, the higher the priority would be, meaning the task should be executed earlier. We made it this way to reflect our personal work habits: we prefer to accomplish longer tasks earlier to feel more relaxed later with shorter tasks knowing that we are meeting all the deadlines. Then, the more dependencies a task has, the less the priority of a task will be. It makes sense because we want to complete the tasks with no (or fewer) dependencies first rather than the tasks that depend on others. 

Mathematically, this function looks like this: 
***priority=e^(duration//60) / (1+e^(duration//60)) - number of dependencies***

Here, we take the duration of a task in minutes, convert it to hours, and then apply the inverse logit function to it. The inverse logit function maps arbitrary duration numbers into the [0,1] range, with higher duration values being mapped closer to 1 (higher priority). This way, we can easier compute and compare priority values across tasks. After we applied the inverse logit function, we subtracted the number of dependencies for that task, which creates an inverse relationship between the number of dependencies and priority values: the bigger the number of dependencies, the smaller the priority value will result, and the other way around.

As a result, we obtain priority values that are on the same scale, can be compared, and perfectly reflect the intuitive reasoning behind which tasks should be executed earlier.

To reflect this priority function in code, we implemented a max-heap structure so that tasks with higher priority were placed at the top of the heap and hence, were executed first.

```python
#The code was adapted from CS110 Session 13
class Task:
    """
    A class to represent a task.
    Attributes:
    - id: Task id (a reference number)   
    - description: Task short description   
    - duration: Task duration in minutes   
    - dependencies: List of task ids that need to preceed this task    
    """
    
    # Initialize an instance of Task
    def __init__(self, id, description, duration, dependencies, time=None, task_type="normal"):
        """
        Initializes an instance of Task.
        
        Parameters:
        ---------
        id: int
            Task id (a reference number)   
        description: str
            Task short description   
        duration: int
            Task duration in minutes   
        dependencies: lst
            List of task ids that need to preceed this task  
        time: int
            The time at which the task is completed (optional)
        task_type: str
            Dummy or not 
        """
        
        # Assign input values to attributes
        self.id = id
        self.description = description
        self.duration = duration
        self.dependencies = dependencies
        self.time = time
        
        # Calculate the priority of the task based on its duration, dependencies, and task type
        if task_type == "normal":
            self.priority = np.exp(self.duration/60)/(1+np.exp(self.duration/60)) - len(self.dependencies)
        else:
            self.priority = -float('inf')
            
    def __lt__(self, alternative_task):
        """
        Determines if the current task has a lower priority than another task.
        
        Parameters
        ---------
        alternative_task: Task() object
            The task to compare to
        
        Returns:
        True if the current task has a lower priority than the alternative task; otherwise False.
        """
        
        # Compare the time or priority of the current task to the alternative task
        if self.time != None and alternative_task.time != None:
            return self.time > alternative_task.time    
        else:
            return self.priority < alternative_task.priority
```

```python
#The code was adapted from CS110 Session 13
class MaxHeapq:
    """ 
    A class that implements properties and methods 
		that support a max priority queue data structure

	  Attributes
	  ----------
	  heap : arr
	      A Python list where key values in the max heap are stored
	  heap_size: int
	      An integer counter of the number of keys present in the max heap
	  """  

    def __init__(self):    
        """
        Parameters
        ----------
        None
        """    
        self.heap       = []
        self.heap_size  = 0
        
    def left(self, i):
        """
        Takes the index of the parent node
        and returns the index of the left child node

        Parameters
        ----------
        i: int
          Index of parent node

        Returns
        ----------
        int
          Index of the left child node

        """
        return 2 * i + 1

    def right(self, i):
        """
        Takes the index of the parent node
        and returns the index of the right child node
        
        Parameters
        ----------
        i: int
            Index of parent node

        Returns
        ----------
        int
            Index of the right child node

        """

        return 2 * i + 2
		
    def parent(self, i):
        """
        Takes the index of the child node
        and returns the index of the parent node
        
        Parameters
        ----------
        i: int
            Index of child node

        Returns
        ----------
        int
            Index of the parent node

        """

        return (i - 1)//2

    def maxk(self):     
        """
        Returns the highest key in the priority queue. 
        
        Parameters
        ----------
        None

        Returns
        ----------
        int
            the highest key in the priority queue

        """
        return self.heap[0]     
    
  
    def heappush(self, dummy, key):  
        """
        Insert a key into a priority queue 
        
        Parameters
        ----------
        dummy: Task()
            A dummy task object to compare to
        key: int
            The key value to be inserted

        Returns
        ----------
        None
        """
        # Append dummy to the heap list
        self.heap.append(dummy)
        
        # Call the increase_key method with heap_size and key as arguments
        self.increase_key(self.heap_size,key)
        
        # Increment heap_size by 1
        self.heap_size+=1
        
    def increase_key(self, i, key): 
        """
        Modifies the value of a key in a max priority queue
        with a higher value
        
        Parameters
        ----------
        i: int
            The index of the key to be modified
        key: int
            The new key value

        Returns
        ----------
        None
        """
        # Check if the new key is smaller than the current key
        # if yes, raise a ValueError exception
        if key < self.heap[i]:
            raise ValueError('new key is smaller than the current key')

        # Update the value of the key at index i to the new key value.
        self.heap[i] = key
        
        # Reorder the heap to maintain the max-heap property
        while i > 0 and self.heap[self.parent(i)] < self.heap[i]:
            
            # If the parent node has a smaller value than the current node
            # then swap the two nodes.
            j = self.parent(i)
            holder = self.heap[j]
            self.heap[j] = self.heap[i]
            self.heap[i] = holder
            i = j    
            
            
       
    def heapify(self, i):
        """
        Creates a max heap from the index given
        
        Parameters
        ----------
        i: int
            The index of of the root node of the subtree to be heapify

        Returns
        ----------
        None
        """
        l = self.left(i)
        r = self.right(i)
        heap = self.heap
        
        # Determine which node has the largest value among the current node,
        # its left child, and its right child
        if l <= (self.heap_size-1) and heap[l]>heap[i]:
            largest = l
        else:
            largest = i
        if r <= (self.heap_size-1) and heap[r] > heap[largest]:
            largest = r

        # If the largest node is not the current node, swap the values of
        # the current node and the largest node, and recursively call heapify
        if largest != i:
            heap[i], heap[largest] = heap[largest], heap[i]
            self.heapify(largest)

    def heappop(self):
        """
        returns the larest key in the max priority queue
        and remove it from the max priority queue
        
        Parameters
        ----------
        None

        Returns
        ----------
        int
            the max value in the heap that is extracted
        """
        if self.heap_size < 1:
            raise ValueError('Heap underflow: There are no keys in the priority queue ')
        # Extract the maximum key
        maxk = self.heap[0]
        # Replace the first element with the last element in the heap
        self.heap[0] = self.heap[-1]
        # Remove the last element in the heap (which is the previous maximum key)
        self.heap.pop()
        # Heapify the modified heap to maintain the max-heap property
        self.heap_size-=1
        self.heapify(0)
        return maxk
```

```python
#The code was adapted from CS110 Session 13
class TaskScheduler:
    """
    A Simple Daily Task Scheduler Using Priority Queues
    """

    def __init__(self, tasks):
        """
        Initializes the TaskScheduler with a list of tasks, and two priority queues

        Parameters
        ---------
          tasks: lst
            A list of tasks to be scheduled
        """
        # Initializes the TaskScheduler with a list of tasks, and two priority queues
        self.tasks = tasks
        self.flexible_priority_queue = MaxHeapq()  
        self.fixed_priority_queue = MaxHeapq()    
        
    def print_self(self):
        """
        Prints the list of tasks added to the scheduler
        """
        # Prints the list of tasks added to the scheduler
        print("Tasks added to the simple scheduler:")
        print("--------------------------------------")
        for t in self.tasks:
            print(f"➡️'{t.description}', duration = {t.duration} mins.")   
            if len(t.dependencies)>0:
                print(f"\t ⚠️ This task depends on others!")     
                 
            
    def get_tasks_ready(self):
        """
        Adds tasks to the priority queues (fixed and flexible) based on their starting time and dependencies
        """
        # Adds tasks to the priority queues (fixed and flexible) based on their starting time and dependencies
        for task in self.tasks:
            if task.time==None:
                self.flexible_priority_queue.heappush(Task(0,'placeholder',-float('inf'),[],task_type = "dummy"), task)
            else:
                self.fixed_priority_queue.heappush(Task(0,'placeholder',-float('inf'),[],float('inf'),task_type = "dummy"), task)

    
    def format_time(self, time):
        """
        Formats time in minutes to a human-readable format (hours and minutes)

        Parameters
        ----------
          time: int
            Time in minutes to be formatted

        Returns
        ----------
          str: A string in the format of hours and minutes
        """
        # Formats time in minutes to a human-readable format (hours and minutes)
        return f"{int(time//60)}h{int(time%60):02d}"


    def run_task_scheduler(self, starting_time):
        """
        Runs the task scheduler starting from the given time (in minutes)

        Parameters
        ----------
          starting_time: int
            The starting time for the scheduler in minutes
        """
        # Runs the task scheduler starting from the given time (in minutes)
        current_time = starting_time
        print("Running a simple scheduler:\n")

        # Adds tasks to the priority queues (fixed and flexible)
        self.get_tasks_ready()

        while self.flexible_priority_queue.heap or self.fixed_priority_queue.heap:
            # Continues until both priority queues are empty

            #take the top tasks from both queues
            if len(self.flexible_priority_queue.heap) > 0:
                top_flexible = self.flexible_priority_queue.maxk()
            else:
                top_flexible = Task(0,'placeholder',float('inf'),[],task_type = "dummy")

            if len(self.fixed_priority_queue.heap) > 0:
                top_fixed = self.fixed_priority_queue.maxk()
            else:
                top_fixed = Task(0,'placeholder',float('inf'),[],float('inf'), task_type = "dummy")

            if current_time+top_flexible.duration > top_fixed.time:
                # If the top flexible task interferes with the top fixed task, complete the fixed task first
                print(f"🕰t={self.format_time(current_time)}")
                print(f"\trelaxing for {top_fixed.time-current_time} mins...")
                current_time=top_fixed.time
                print(f"\t✅ t={self.format_time(current_time)}, task completed!")
                execute_task = self.fixed_priority_queue.heappop()
            else:
                # Otherwise execute the flexible task
                execute_task = self.flexible_priority_queue.heappop()

            print(f"🕰t={self.format_time(current_time)}")
            print(f"\tstarted '{execute_task.description}' for {execute_task.duration} mins...")
            current_time += execute_task.duration            
            print(f"\t✅ t={self.format_time(current_time)}, task completed!") 

        total_time = current_time - starting_time             
        print(f"\n🏁 Completed all planned tasks in {total_time//60}h{total_time%60:02d}min!")
```

```python
def create_tasks(tasks_dict):
    '''
    Creates a list of Task objects from a dictionary of tasks.
    
    Parameters
    ----------
    tasks_dict: lst
         A list of dictionaries, where each dictionary represents a task.
    
    Returns
    ----------
    tasks_objects: lst
        A list of Task objects created from the given list of dictionaries.   
    '''
    tasks_objects=[]
    for i in tasks_dict:
        tasks_objects.append(Task(i['id'],i['description'],i['duration'],i['dependencies'],i['time']))

    return tasks_objects
```

**Theoretical Time Complexity Analysis**<br>
**(`class TaskScheduler` methods)**

1. `init(self, tasks)`

  - Recurrence relation: T(n) = T(n-1) = O(1)
  - Time complexity: **O(1)** 
  - Interpretation: The algorithm runtime does not scale and stays as a constant with the increase of the input size n.
  - Explanation: The method initializes the `TaskScheduler` object with the given tasks, and initializes two priority queues, which take constant time to create. 
  
2. `print_self(self)`

  - Recurrence relation: T(n) = T(n-1) + O(1)
  - Time complexity: **O(n)**
  - Interpretation: The algorithm runtime scales linearly with the increase of the input size n.
  - Explanation: The method prints each task in the list of tasks, so the time complexity is linear in the number of tasks in the list. 

3. `get_tasks_ready(self)`

  - Recurrence relation: T(n) = T(n-1) + log n (each layer, we have log n)
  - Time complexity: **O(nlogn)**
  - Interpretation: The algorithm runtime scales in the order of n*logn with the increase of the input size n.
  - Explanation: The method adds tasks to the priority queues based on their starting time and dependencies. For each task, it performs either one heappush operation on the fixed priority queue or one heappush operation on the flexible priority queue. Each heappush operation takes O(logn) time in the worst case, so the overall time complexity is O(nlogn). 

4. `format_time(self, time)`

  - Recurrence relation: T(n) = T(n-1) = O(1)
  - Time complexity: **O(1)**
  - Interpretation: The algorithm runtime does not scale and stays as a constant with the increase of the input size n.
  - Explanation: The method simply formats the given time into a string, which takes constant time.

5. `run_task_scheduler(self, starting_time)`

  - Recurrence relation: T(n) = T(n-1) + log n (each layer, we have log n)
  - Time complexity: **O(nlogn)**
  - Interpretation: The algorithm runtime scales in the order of n*logn with the increase of the input size n.
  - Explanation: The method runs the task scheduler starting from the given time. It repeatedly performs one of two operations until both priority queues are empty: either one heappop operation on the fixed priority queue or one heappop operation on the flexible priority queue. Each heappop operation takes O(logn) time in the worst case. In addition, the method calls the get_tasks_ready method, which has a time complexity of O(nlogn). Therefore, the overall time complexity is O(nlogn)


**Experimental Time Complexity Analysis**<br>

We designed an experiment below to verify the theoretical time complexity analysis of the `run_task_scheduler` method, as this is the **essential** method that we care about. The experiment measures the average runtime of a task scheduler algorithm for different input sizes from 1 to `max_tasks = 100`.

The `generate_tasks` function generates a list of task objects with random dependencies. For each task, it generates a list of dependencies by randomly selecting tasks from earlier in the list. The duration of each task is also randomly generated.

The main loop in the code iterates over different input sizes (the number of tasks), and for each input size, it runs the task `scheduler algorithm` 500 times to obtain an average runtime. 

The average runtime for each input size is calculated as the average of the runtimes across multiple experiments. The results are plotted as a scatter plot of input size versus average runtime. The plot shows how the runtime of the task scheduler algorithm increases as the number of tasks increases.





```python
import tqdm 

def generate_tasks(num_tasks):
    """
    Generate a list of task objects with random dependencies.

    Parameters
    ----------
        num_tasks: int
            The number of tasks to generate.

    Returns
    ---------
        list: A list of task objects with the following properties:
            id (int): The task's ID.
            description (str): The task's description.
            duration (int): The task's duration in minutes.
            dependencies (list): A list of IDs of tasks that this task depends on.
            time (int): The time at which this task is completed.
    """
    tasks = []
    for i in range(num_tasks):
        dependencies = []
        for j in range(i):
            if random.random() < 0.3:
                dependencies.append(j)
        tasks.append({
            'id': i,
            'description': f'Task {i}',
            'duration': random.randint(10, 120),
            'dependencies': dependencies,
            'time': None
        })
    return tasks

max_tasks = 100
avg_runtime = []
input_sizes = list(range(1,max_tasks+1))
start_scheduler_at = 8*60

experiments = 500
for input_size in tqdm.tqdm(input_sizes):
    runtime = []
    for _ in range(experiments):

        # Generate random tasks and create task objects
        random_inputs = generate_tasks(input_size)
        tasks_objects = create_tasks(random_inputs)
        task_scheduler = TaskScheduler(tasks_objects)

        # Run the task scheduler and measure the time taken
        time_start = time.time()
        blockPrint()
        task_scheduler.run_task_scheduler(start_scheduler_at)
        enablePrint()
        runtime.append(time.time()-time_start)

    # Calculate the average runtime for this input size
    avg_runtime.append(sum(runtime)/len(runtime))

# Plot the results
plt.figure()
plt.plot(input_sizes, avg_runtime)
plt.title("Task Scheduler Average Runtime")
plt.xlabel("The Number of Tasks n")
plt.ylabel("Average Time (seconds)")
plt.show()
```

The x-axis demonstrates the number of tasks (input size n), and the y-axis shows the average runtime of 50 experiments at each input size in seconds. 

As shown by the plot, with the increase of the number of tasks n, the average runtime scales in an order of O(n logn), as predicted in the theoretical time complexity analysis. The shape generally matches despite some minor discrepancies around n = 70.
